```python
import numpy as np
import pandas as pd
```

```python
music = pd.read_csv("1990sClassicHits.csv")
```

```python
music.head(5)
```

```python
music.describe()
```

```python
music.isna().any()
```

```python
def convert_duration(duration):
    minutes, seconds = map(int, duration.split(':'))
    return minutes * 60 + seconds

music['Duration'] = music['Duration'].apply(convert_duration)
```

```python
music
```

```python
music.to_csv('90sHits.csv', index = False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_music = music.select_dtypes(include=[np.number])

# Compute the correlation matrix
correlation_matrix = numeric_music.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('90s Hits Correlation Matrix')
plt.show()
```
